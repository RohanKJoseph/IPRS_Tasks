#Hands-on Study of Imputation Methods in Health Data

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. creating the dummy datas..
data = {
    "Patient_ID": ["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10"],
    "Age" : [25,30,np.nan,22,40,np.nan,29,31,28,np.nan],
    "Blood_Pressure" : [120,130,125,np.nan,140,135,np.nan,128,132,130],
    "Gender" : ["M","F","F","M",np.nan,"F","M","M",np.nan,"F"],
    "Cholesterol" : [200,220,210,205,np.nan,215,225,np.nan,230,210]
}

#creating the data frame

df = pd.DataFrame(data)

#displaying the dataframe
print("Data Frame :\n",df)
```

## Inspect the Dataset

```python
# checking data types

df.info()
```

```python
# checking missing values

print(df.isnull().values.any())
print("\n")
print("Count of missing values per column :")
print("*"*55)
print(df.isnull().sum())

print("\n\n")

print("Percentage of missing values per column :")
print("*"*55)
print((df.isnull().sum() / len(df)) * 100)
```

```python
#numerical statistics
print("Numerical Statistics:")
df.describe()
```

## Impute Missing Values

```python
# Create copies to do eact type of impute
df_mean = df.copy()
df_median = df.copy()
numerical_cols = ['Age', 'Blood_Pressure', 'Cholesterol']
```

### 1. Mean Imputation:

```python

for col in numerical_cols:
    df_mean[col] = df_mean[col].fillna(df_mean[col].mean())
```

###2. Median Imputation

```python
for col in numerical_cols:
    df_median[col] = df_median[col].fillna(df_median[col].median())
```

### 3. Mode Imputation

```python
gender_mode = df['Gender'].mode()[0]   # [0] grabs the first mode value if there's a tie
df_mean['Gender'] = df_mean['Gender'].fillna(gender_mode)
df_median['Gender'] = df_median['Gender'].fillna(gender_mode)
```

```python
#checking the data frames after the imputation..
print("Mean Imputed Data Frame :\n",df_mean)
print(df_mean.isnull().sum())
print("\n")
print("Median Imputed Data Frame :\n",df_median)
print(df_median.isnull().sum())

#from the output its clear that there is no missing values in both the mean  and median data frames
```

## Compare Effects

```python
#displaying the data summary of original data
print("ORIGINAL DATA SUMMARY ")
print("="*50)
print(df[numerical_cols].describe())

#displaying the data summary of mean imputation summary
print("\n MEAN IMPUTATION SUMMARY ")
print("="*50)
print(df_mean[numerical_cols].describe())

#displaying the data of median imputation summary
print("\n MEDIAN IMPUTATION SUMMARY ")
print("="*50)
print(df_median[numerical_cols].describe())
```
