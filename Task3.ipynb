##Converting categorical variables into numerical form (**encoding**)

```python
#converting the date to a numeric value finding the waiting time
#which is by subtracting the AppointmentDay from ScheduledDay
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

df['WaitingDays'] = (
    df['AppointmentDay'] - df['ScheduledDay']
).dt.days

# print(df.WaitingDays)
```

```python
#Encode Categorical Variables

df.select_dtypes(include='object').columns #Identify Categorical Columns
```

```python
#1 Label encoding for columns like Gender and No-Show
#as we can use binary variables

df['Gender'] = df['Gender'].map({'F':0, 'M':1})
df['No-show'] = df['No-show'].map({'No':1, 'Yes':0})
```

```python
#2 One- hot encoding for multy catagory columns like Neighbourhood

df = pd.get_dummies(df,columns=['Neighbourhood'],drop_first=True)
```

```python
#checking the values of age
print(df['Age'].min())
print(df['Age'].max())

#removing the invalid ages
df = df[df['Age'] >= 0]

print(df['Age'].min())  #now the age is varified
```

```python
#Check Binary Medical Condition Columns

binary_cols = [
    'Scholarship',
    'Hipertension',
    'Diabetes',
    'Alcoholism',
    'Handcap',
    'SMS_received'
]

for col in binary_cols:
    print(col, df[col].unique())
```
