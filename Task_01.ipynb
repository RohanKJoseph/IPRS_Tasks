# Data Preprocessing of Student Dataset

```python
import pandas as pd  #aliyasing it as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

## Understand the dataset structure

```python
#1 Load the CSV file using Pandas
data = pd.read_csv("/content/student_data_100_preprocess.csv")
```

```python
#2 Display first 5 rows

print("First 5 rows")
print(data.head(5))
print("\n")

#Displaying the last 5 rows
print("Last 5 rows")
print(data.tail(5))

```

```python
#3 Shape of the dataset

print("Shape of the Data set is : ",data.shape)
print("\n")

#Column names

print("Column Names are : ",data.columns.tolist()) # converting into list format
print("\n")

#Data types

print("Data types are as follows\n ",data.dtypes)
```

```python
#4 Identify neumerical columns

numeric_columns = data.select_dtypes(include=['number']).columns.tolist()
print("Numerical columns are : ", numeric_columns)
print("\n")

#identify categorical columns

categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns are : ", categorical_columns)
print("\n")

```

```python
#5  statistical summary of numerical columns

print("statistical summary of numerical columns are as follows")
print("="*55)
print(data.describe())
```

## Identify Missing Values

```python
#counting the missing values per column

print("No of missing values are as follows")
print("="*35)
print(data.isnull().sum())
```

```python
#identifying which columns contain missing data

print("Columns with missing values are : ",data.columns[data.isnull().any()].tolist())
```

```python
#Deciding which column needs Mean filling and Mode filling

#giving Mean Filling to numerical columns
for i in numeric_columns:
  if data[i].isnull().any():
    data[i] = data[i].fillna(data[i].mean())

#giving Mode Filling to categorical columns
for j in categorical_columns:
    if data[j].isnull().any():
        data[j] = data[j].fillna(data[j].mode()[0])
```

```python
#varifying the filling

print("No of missing values are as follows")
print("="*35)

print(data.isnull().sum())
```

```python
print("The data info")
print("="*55)
print(data.info())
print("\n")

print("The data description")
print("="*55)
print(data.describe())
```

## Encode Categorical Variables

```python
#Encode Gender using Label Encoding
if 'Gender' in data.columns:
    le = LabelEncoder()
    data['Gender'] = le.fit_transform(data['Gender'])

#Apply One-Hot Encoding to Parental_Education using Pandas
if 'Parental_Education' in data.columns:
    data = pd.get_dummies(data, columns=['Parental_Education'])
```

```python
#  Confirm dataset contains only numerical values

print("The data info")
print("="*55)
print(data.info())
print("\n")

```

```python
non_numeric_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()
print("Non-numeric columns in the dataset:", non_numeric_columns)
```

## Prepare Data for Model Training

```python
#Separate data to features: x and target: y

X = data.drop(columns=['Final_Score'])
y = data['Final_Score']

#splitting the dataset in 80% for training and 20% for testing

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20,random_state=42
)
```

```python
#split proportions

print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape:  {X_test.shape}")
print(f"Training target shape:   {y_train.shape}")
print(f"Testing target shape:    {y_test.shape}")
```
