#Transforming Categorical Data into Numerical Format

```python
#importing therequired libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#creating the dummy data set and dataframe

data = {
    'PAtient_ID': ["P1", "P2", "P3", "P4", "P5"],
    'Gender': ["M","F","F","M","F"],
    "AppointmentType" : ["Routine","Emergency","Routine","Follow-up","Emergency"],
    "City":["New York","Los Angeles","Chicago","New York","Chicago"]
}

df = pd.DataFrame(data)
print("Original Data set : \n",df)
```

##Inspect the Dataset

```python
df.info()
print("\n")
df.describe() #to check for the numeric statistics
```

```python
#checking for missing values
df.isnull().sum()

#from its output its clear that there is no missing vlues
```

```python
# creating the copy of the data for applying both of the operation

df_label = df.copy()
df_onehot = df.copy()
```

##Label Encoding

```python
#Using LabelEncoder we encode each categorical column

#initialising the labelencoder

le = LabelEncoder()

#applying the label encoder on the catogarical columns

catogarical_columns = df_label.select_dtypes(include=['object']).columns #finding the catogarical clms to encode
#applying on each
df_label[catogarical_columns] = df_label[catogarical_columns].apply(le.fit_transform)

print("Label Encoded Data : \n",df_label) #displaying the output
```

##One-Hot Encoding

```python
#Using one hot encoding we encode each categorical column

categorical_cols = ['Gender', 'AppointmentType', 'City']

#applying the one hot encoder on the catogarical columns using the dummies
df_onehot = pd.get_dummies(df, columns=categorical_cols , dtype=int)

print("--- One-Hot Encoded Dataset ---")
print(df_onehot)   #displying the output
```
